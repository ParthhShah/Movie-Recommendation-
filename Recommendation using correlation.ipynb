```python
# import necessary libraries
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel

# load the dataset
movies = pd.read_csv('movies.csv')
ratings = pd.read_csv("ratings.csv")

print(movies.head())
ratings.head()
```

```python
data = ratings.merge(movies,on='movieId', how='left')
```

```python
data
```

# Calculating the average of the rating

```python
Average_ratings = pd.DataFrame(data.groupby('title')['rating'].mean())
Average_ratings.head(10)
```

# Counting The no. of ratings done by the user for each movie

```python
Average_ratings['Total Ratings'] = pd.DataFrame(data.groupby('title')['rating'].count())
Average_ratings.head(10)
```

# Converting this into matrix using .pivot

```python
movie_user = data.pivot_table(index='userId',columns='title',values='rating')
```

```python
movie_user
```

# Here we are using Corrwith which works as correlation and give us the result

```python
a="Toy Story (1995)"
correlations = movie_user.corrwith(movie_user[a])
correlations.head(10)
```

# Problem with this is we get alot of null values thus removing the null values 

```python
recommendation = pd.DataFrame(correlations,columns=['Correlation'])
recommendation.dropna(inplace=True)
recommendation = recommendation.join(Average_ratings['Total Ratings'])
recommendation.head()
```

# sorting all the recommendation in descending 

```python
recc = recommendation[recommendation['Total Ratings']>10].sort_values('Correlation',ascending=False).reset_index()
```

```python
recc = recc.merge(movies,on='title', how='left')
print('The recommandation  for the movie:',a)
recc.head(10)
```
